```python
import pandas as pd
import csv
import numpy as np
import seaborn as sns

dataset_hr = pd.read_excel("D:\Python\Codex\dataset HR\dataset HR analytics.xlsx")
print(dataset_hr.info())
print(dataset_hr.shape)
dataset_hr.head(10)
```

```python
# ===== Convert Scan Record, data type as datetime =====

dataset_hr["DOB"] = pd.to_datetime(dataset_hr["DOB"], format='%m/%d/%Y')
dataset_hr["DateofHire"] = pd.to_datetime(dataset_hr["DateofHire"], format='%m/%d/%Y')
dataset_hr["DateofTermination"] = pd.to_datetime(dataset_hr["DateofTermination"], format='%m/%d/%Y')

dataset_hr.head()
print(dataset_hr.info())
print(dataset_hr.shape)
```

```python
dataset_hr["umur"] = (pd.Timestamp.now().normalize() - dataset_hr["DOB"]).astype('<m8[Y]')

print(dataset_hr.info())
print(dataset_hr.shape)
```

```python
dataset_hr["umur"] = dataset_hr["umur"].astype(int)
print(dataset_hr.info())
```

```python
#Remove unneeded columns

#Remove the column EmployeeNumber
datahr_corr = dataset_hr.drop(['EmpID','DOB','DateofHire','DateofTermination',
                               'Employee_Name','LastPerformanceReview_Date'], axis = 1)
datahr_corr.info()
```

```python
#Transform non-numeric columns into numerical columns
from sklearn.preprocessing import LabelEncoder

for column in datahr_corr.columns:
        if datahr_corr[column].dtype == np.number:
            continue
        datahr_corr[column] = LabelEncoder().fit_transform(datahr_corr[column])
datahr_corr.info()
```

```python
#Get the correlation of the columns
datahr_corr.corr()
```

```python
#Visualize the correlation
plt.figure(figsize=(14,14))  #14in by 14in
sns.heatmap(datahr_corr.corr(), annot=True, fmt='.0%')
```

```python
#Split the data into independent 'X' and dependent 'Y' variables
X = datahr_corr.iloc[:, 1:datahr_corr.shape[1]].values 
Y = datahr_corr.iloc[:, 0].values
```

```python
# Split the dataset into 75% Training set and 25% Testing set
from sklearn.model_selection import train_test_split
X_train, X_test, Y_train, Y_test = train_test_split(X, Y, test_size = 0.25, random_state = 0)
```

```python
#Use Random Forest Classification algorithm
from sklearn.ensemble import RandomForestClassifier
forest = RandomForestClassifier(n_estimators = 10, criterion = 'entropy', random_state = 0)
forest.fit(X_train, Y_train)
```

```python
#Get the accuracy on the training data
forest.score(X_train, Y_train)
```

```python
#Show the confusion matrix and accuracy for  the model on the test data
#Classification accuracy is the ratio of correct predictions to total predictions made.
from sklearn.metrics import confusion_matrix

cm = confusion_matrix(Y_test, forest.predict(X_test))
  
TN = cm[0][0]
TP = cm[1][1]
FN = cm[1][0]
FP = cm[0][1]
  
print(cm)
print('Model Testing Accuracy = "{}!"'.format(  (TP + TN) / (TP + TN + FN + FP)))
print()# Print a new line
```

```python
# Return the feature importances (the higher, the more important the feature).
importances = pd.DataFrame({'feature':datahr_corr.iloc[:, 1:datahr_corr.shape[1]].columns,'importance':np.round(forest.feature_importances_,3)}) #Note: The target column is at position 0
importances = importances.sort_values('importance',ascending=False).set_index('feature')
importances
```

```python
#Visualize the importance
importances.plot.bar()
```
